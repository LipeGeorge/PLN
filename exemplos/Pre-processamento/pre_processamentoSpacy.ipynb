# 1. Carregar o modelo de português

```python
import spacy

nlp = spacy.load("pt_core_news_sm")

texto = "Tatiana Sampaio, pesquisadora da UFRJ, desenvolveu a polilaminina, proteína que recuperou movimentos de pacientes paraplégicos/tetraplégicos em estudos. A Anvisa autorizou ensaios clínicos em 2026 para tratar lesões na medula."
```

# 2. Segmentação em sentenças

```python
doc = nlp(texto)

for sent in doc.sents:
    print(sent.text)
```

# 3. Tokenização

```python
doc = nlp(texto)

tokens = [token.text for token in doc]

print(tokens)
```

#4. Remoção de pontuação

```python
doc = nlp(texto)

tokens_sem_pontuacao = [token.text for token in doc if not token.is_punct]

print(tokens_sem_pontuacao)
```

#5. Remoção de stopwords

```python
doc = nlp(texto)

tokens_sem_stopwords = [token.text for token in doc if not token.is_stop and not token.is_punct]

print(tokens_sem_stopwords)
```

# 6. Lematização

```python
doc = nlp(texto)

lemmas = [token.lemma_ for token in doc]

print(lemmas)
```

# 7. NER

```python
doc = nlp(texto)

tokens_processados = [
    token.lemma_
    for token in doc
    if not token.is_stop and not token.is_punct
]

print(tokens_processados)
```

# Pipeline final de pré-processamento

```python
doc = nlp(texto)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
