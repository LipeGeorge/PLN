# Texto de exemplo


```python


texto = """
O processamento da linguagem natural permite que os computadores
entendam e analisem textos escritos por humanos.
"""


```

# 1. normalização

```python
texto = texto.lower()
```

#  2.1. Tokenização com expansão de contração

```python
contracoes = {

    # de + artigos
    "do": ["de", "o"],
    "da": ["de", "a"],
    "dos": ["de", "os"],
    "das": ["de", "as"],

    # de + artigos indefinidos
    "dum": ["de", "um"],
    "duma": ["de", "uma"],
    "duns": ["de", "uns"],
    "dumas": ["de", "umas"],

    # de + demonstrativos
    "deste": ["de", "este"],
    "desta": ["de", "esta"],
    "destes": ["de", "estes"],
    "destas": ["de", "estas"],

    "desse": ["de", "esse"],
    "dessa": ["de", "essa"],
    "desses": ["de", "esses"],
    "dessas": ["de", "essas"],

    "daquele": ["de", "aquele"],
    "daquela": ["de", "aquela"],
    "daqueles": ["de", "aqueles"],
    "daquelas": ["de", "aquelas"],

    # de + pronomes neutros
    "disto": ["de", "isto"],
    "disso": ["de", "isso"],
    "daquilo": ["de", "aquilo"],


    # em + artigos
    "no": ["em", "o"],
    "na": ["em", "a"],
    "nos": ["em", "os"],
    "nas": ["em", "as"],

    # em + indefinidos
    "num": ["em", "um"],
    "numa": ["em", "uma"],
    "nuns": ["em", "uns"],
    "numas": ["em", "umas"],

    # em + demonstrativos
    "neste": ["em", "este"],
    "nesta": ["em", "esta"],
    "nestes": ["em", "estes"],
    "nestas": ["em", "estas"],

    "nesse": ["em", "esse"],
    "nessa": ["em", "essa"],
    "nesses": ["em", "esses"],
    "nessas": ["em", "essas"],

    "naquele": ["em", "aquele"],
    "naquela": ["em", "aquela"],
    "naqueles": ["em", "aqueles"],
    "naquelas": ["em", "aquelas"],

    # em + pronomes neutros
    "nisto": ["em", "isto"],
    "nisso": ["em", "isso"],
    "naquilo": ["em", "aquilo"],


    # a + artigos
    "ao": ["a", "o"],
    "aos": ["a", "os"],
    "à": ["a", "a"],
    "às": ["a", "as"],

    # a + demonstrativos
    "àquele": ["a", "aquele"],
    "àquela": ["a", "aquela"],
    "àqueles": ["a", "aqueles"],
    "àquelas": ["a", "aquelas"],

    # a + pronome neutro
    "àquilo": ["a", "aquilo"],


    # por + artigos
    "pelo": ["por", "o"],
    "pela": ["por", "a"],
    "pelos": ["por", "os"],
    "pelas": ["por", "as"]
}
```

```python
tokens_expandidos = []
for token in tokens:
    if token in contracoes:
        tokens_expandidos.extend(contracoes[token])
    else:
        tokens_expandidos.append(token)
```

```python
tokens_expandidos
```

# 3. remoção de pontuação

```python
tokens = [t for t in tokens if t not in string.punctuation]
```

```python
tokens
```

# 4. remoção de stopwords

```python

stop_words = set(stopwords.words('portuguese'))
tokens = [t for t in tokens if t not in stop_words]
```

```python
tokens
```

# 5. stemming

```python

stemmer = RSLPStemmer()
tokens_stem = [stemmer.stem(t) for t in tokens]
```

```python
tokens_stem
```

# 6. lematizador

A NLTK possui suporte limitado para lematização em português. O lematizador nativo da biblioteca (WordNetLemmatizer) foi desenvolvido para inglês, portanto, para português normalmente são usadas três abordagens:

- usar o WordNetLemmatizer (exemplo didático)

- usar dicionários de lemas externos

- usar bibliotecas mais adequadas como spaCy ou Stanza

Abaixo está um exemplo simples de lematização usando NLTK, apenas para demonstrar o processo.

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# baixar recursos necessários
nltk.download('wordnet')
tokens2 = ['os', 'estudantes', 'estavam', 'estudando', 'e', 'aprenderam', 'muitos', 'conceitos', '.']
lemmas = [lemmatizer.lemmatize(token) for token in tokens2]
```

```python
print("Tokens:", tokens2)
print("Lemas:", lemmas)
```

# 7. Setenciador

```python
from nltk.tokenize.punkt import PunktSentenceTokenizer

texto = "A pesquisa começou em 2020. Os resultados foram publicados em 2024."

tokenizer = PunktSentenceTokenizer()

sentencas = tokenizer.tokenize(texto)

print(sentencas)
```
